```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

# Read data from .dat file
data = np.loadtxt('/home/mfi/Desktop/mfi/full_simulations/2_Guest/rmsd-2d_2guest.dat')

#Set the output frequency defined earlier during the RMSD routine
outputfrequency = 50

# Extract frames and RMSD values
frames = data[:, :2]
rmsd = data[:, 2]
rmsd_max = rmsd.max()


#Get the number of steps from frames
nsteps = int(np.sqrt(len(frames)))

#Make a matrix of size steps x steps
rmsd = rmsd.reshape(nsteps, nsteps)

#Initialize the figure
plt.figure(figsize=(10, 10))

# Adjust the color map and normalization
norm = Normalize(vmin=0, vmax=rmsd_max)  # Adjust vmin and vmax as needed
plt.imshow(rmsd, cmap='Greys', interpolation='bicubic', aspect='auto', norm=norm)
plt.colorbar().set_ticks(np.arange(0, rmsd_max, 0.2))

# Calculate tick positions and labels
num_ticks = 5
tick_positions = np.linspace(0, nsteps-1, num_ticks, dtype=int)
#Transfer steps to nanoseconds. Every frame being output equals to 4 fs
tick_labels = np.round(tick_positions * outputfrequency * 4 / 1000000, 2)

plt.xticks(tick_positions, tick_labels)
plt.yticks(tick_positions, tick_labels)
plt.gca().invert_yaxis()
plt.xlabel('Simulation time (ns)')
plt.ylabel('Simulation time (ns)')
plt.title('2D-RMSD')

#save the plot
#plt.savefig("/home/mfi/Desktop/mfi/full_simulations/2DRMSD_Plots/1Guest@MOF_2DRMSD.png", dpi=300)

plt.show()


```
